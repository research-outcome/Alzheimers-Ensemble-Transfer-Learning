```python
data_814 = pd.read_csv(ROOT+'814.csv')
data = pd.read_csv(ROOT+'MY_ADNI_CLEAN.csv')
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_814[['Subject']].values, data_814[['Group']].values, test_size=0.20, random_state=13 )
```

```python
def flatten(t):
    return [item for sublist in t for item in sublist]
```

```python
training = flatten(x_train.tolist())
testing = flatten(x_test.tolist())
```

```python
data.loc[data['Subject'].isin(training)].to_csv( ROOT+'training.csv', index=False, encoding='utf-8-sig')
data.loc[data['Subject'].isin(testing)].to_csv( ROOT+'testing.csv', index=False, encoding='utf-8-sig')
```
